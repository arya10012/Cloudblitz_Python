## **Simple Pattern Printing**

```python
for i in range(1,11):
  print(i)
```

```python
for i in range(10):
  print("*",end="")
```

```python
i=1
while i < 6:
  print(i*"*")
  i=i+1
```

```python
row =5
for i in range(row):
  for j in range(i+1):
    print("*",end="")
  print()
```

```python
row =5
for i in range(row):
  for j in range(row-i):
    print("*",end="")
  print()
```

```python
row=5
for i in range(1,row+1):
    for j in range(1,i+1):
        print(j,end=" ")
    print("")
```

```python
row=5
for i in range(row):
  for j in range(row-i-1):
    print(" ",end="")
  for k in range(2*i+1):
    print("*",end="")
  print()
```
