```python
# Basic Hybrid Inheritance Example

class A:
    def method_a(self):
        print("Method from class A")

class B(A):
    def method_b(self):
        print("Method from class B")

class C(A):
    def method_c(self):
        print("Method from class C")

class D(B, C):
    def method_d(self):
        print("Method from class D")

# Create an instance of class D and call methods from all parent classes
d_instance = D()
d_instance.method_a()
d_instance.method_b()
d_instance.method_c()
d_instance.method_d()
```

```python
#Multiple Inheritance - Python
class wholesaler:
  def whole(self):
    print("I am a wholesaler")

class retailer:
  def retail(self):
    print("I am a retailer")

class distributor(wholesaler,retailer):
  def distributor(self):
    print("I am a distributor")
```

```python
#Hierachial Inheritance - Python
class Milk:
  def milk(self):
    print("I am a milk product")

class cheese(Milk):
  def cheese(self):
    print("I am a cheese product")

class butter(Milk):
  def butter(self):
    print("I am a butter product")

class curd(Milk):
  def curd(self):
    print("I am a curd product")
```

```python
import re

def is_valid_phone_number_with_country_code(phone_number):
    """
    Checks if a phone number has a country code and is in a valid format.

    Args:
        phone_number: The phone number string to validate.

    Returns:
        True if the phone number is valid, False otherwise.
    """
    # Regular expression to match country code (1-3 digits) followed by a hyphen
    # and then a phone number (7-15 digits)
    pattern = re.compile(r'^\d{1,3}-\d{7,15}$')

    if pattern.match(phone_number):
        return True
    else:
        return False

# Example usage:
print(is_valid_phone_number_with_country_code("91-8484078886"))
print(is_valid_phone_number_with_country_code("1-555-1234567"))
print(is_valid_phone_number_with_country_code("8484078886")) # Missing country code
print(is_valid_phone_number_with_country_code("91-123")) # Too short
```
